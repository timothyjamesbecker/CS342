## 3.2 <u>Standard Notations and Common Functions<u>

The texts makes use of many common math notation conventions, but has some that are specific to CS

#### <u>Monotonicity</u>

$f(n)$ is <b>monotonically increasing</b> if $m \leq n \implies f(m) \leq f(n)$

$f(n)$ is <b>monotonically decreasing</b> if $m \geq n \implies f(m) \geq f(n)$

$f(n)$ is <b>strictly increasing</b> if $m < n \implies f(m) < f(n)$

$f(n)$ is <b>strictly decreasing</b> if $m > n \implies f(m) > f(n)$

#### <u>Floor/Ceiling</u>

$\forall x \in \mathbb{R}: x-1 < \lfloor x \rfloor \leq x \leq \lceil x \rceil < x+1$

$\forall n \in \mathbb{Z}: \lceil n/2 \rceil + \lfloor n/2 \rceil = n$

$\forall x \in \mathbb{R} \wedge x \geq 0, \forall a,b \in \mathbb{Z} \wedge a,b>0:$

$\left\lceil \frac{\lceil x/a \rceil}{b} \right\rceil = \left\lceil \frac{x}{ab} \right\rceil, \;\;\;\; \left\lfloor \frac{\lfloor x/a \rfloor}{b} \right\rfloor = \left\lfloor \frac{x}{ab} \right\rfloor$

$\left\lceil \frac{a}{b} \right\rceil \leq \frac{a+(b-1)}{b}, \;\;\;\; \left\lfloor \frac{a}{b} \right\rfloor \geq \frac{a-(b-1)}{b}$

```python
#instance code----------------------------------------------------------------------------------------------
import numpy as np                                                       #import numpy package
x = np.random.sample()*100                                               #0.0 to < 100.0
n = int(np.random.sample()*100)+100                                      #random positive n value 100 to 200
print(x)
print(np.floor(x))
print(np.ceil(x))
```

#### <u>Modular Arithmetic</u>  (aka % in python)

$\forall a \in \mathbb{Z}, \forall b \in \mathbb{Z}^{+}: a \bmod b$ is the remainder of $a/b$

$a \bmod b = a-b \lfloor a/b \rfloor ,\;\;\;\; 0 \leq a \bmod  b < b$

```python
a = 132
b = 5
print(a%b)                    #here is our modulus operator
print(a-b*int(np.floor(a/b))) #here is a regular divide operation casted to an int type
```

#### <u>Logarithms</u>

$lg\;n = log_2\;n$, $\;\;ln\;n = log_e\;n$, $\;\;lg^k\;n = (lg\;n)^k$, $\;\;lg\;lg\;n = lg(lg\;n)$

$\forall a,b,c>0 \in \mathbb{R}: a = b^{log_ba}$, $\;\;log_c(ab) = log_c(a)+log_c(b)$, $\;\;log_b(a^n) = nlog_b(a)$, 

$\;\;log_b(a) = \frac{log_c(a)}{log_c(b)}$, $\;\;log_b(1/a) = -log_b(a)$, $\;\;log_b(a) = \frac{1}{log_a(b)}$, $\;\;a^{log_b(c)} = c^{log_b(a)}$  

Note: $lg\;n = log_2\;n$ is used for the regular log in the text because dividing inputs by 2 is so common.

We say that a function is <b>polylog</b> bound if $f(n) = O(lg^k\;n):\;\;$ constant $k \implies lg^b\;n = o(n^a), \;\;\forall a,b \in \mathbb{R}, a>0$

```python
import matplotlib.pyplot as plt
c = 5
n = np.linspace(1,20,20)  
f1 = n-c                                   
plt.plot(n,f1,label='$f1(n)=n-c$')
f2 = np.log(n)/np.log(2)                                 
plt.plot(n,f2,label='$f2(n)=lg\;n$')
plt.title('$f2(n) = o(f1(n))$');
plt.legend();plt.show()
```

#### <u>Polynomials</u>

$\forall d \in \mathbb{Z}^{+}:$ a poly in $n$ of degree $d$ is a function $p(n)$:

$\displaystyle p(n) = \sum_{i=0}^{d}a_in^i :\;\; \text{constants}\;\; a_0,a_1,...a_d$ (also known as poly coefficiants)

for an asymptotically positive poly $p(n)$ of degree $d$:  $p(n) = \Theta(n^d)$

we say $f(n)$ is <b>poly</b> bound if $f(n) = O(n^k):\; \text{constant}\;k$

```python
c = 5
n = np.linspace(1,15,15)                      
f1 = c*n**2                                   
plt.plot(n,f1,label='$f1(n)=n^2$')
f2 = n**4-2*c*n**3+c*n                                  
plt.plot(n,f2,label='$f2(n)=n^7+cn$')
f3 = n**2.3-2*c*n**2                                      
plt.plot(n,f3,label='$f3(n)=n$')
plt.title('$f2(n) = o(f1(n))$');
plt.legend();plt.show()
```

#### <u>Exponentials</u>

$\forall a,n,m \in \mathbb{R}, \; a>0:$ $\;\;a^0 = 1$, $\;\;a^1 = a$, $\;\;a^-1 = \frac{1}{a}$, $\;\;(a^m)^n = a^{mn}$, $(a^n)^m = (a^m)^n$, $\;\;a^ma^n = a^{m+m}$

$\lim\limits_{n \to \infty} \frac{n^b}{a^n} = 0 \implies n^b = o(a^n)$

```python
a = 1.132
b = 7
n = np.linspace(1,15,15)                      
f1 = n**b                                  
plt.plot(n,f1,label='$f1(n)=n^7$')
f2 = a**n                            
plt.plot(n,f2,label='$f2(n)=a^n$')
plt.title('$f1(n) = o(a^n)$'+'  ?');
plt.legend();plt.show()
```

#### <u>Factorials</u>

$\displaystyle n! = 1\cdot2\cdot3\cdot \;...\; n = \prod_{i=1}^{n}i = 
\begin{cases} 
   1 & \text{if } n = 0 \\
   n(n-1)! & \text{if } n > 0
\end{cases}$

```python
import math
n = np.linspace(1,15,15)                      
f1 = 8**n                                 
plt.plot(n,f1,label='$f1(n)=2^n$')
f2 = [math.factorial(n[i]) for i in range(len(n))]                          
plt.plot(n,f2,label='$f2(n)=n!$')
plt.title('$f2(n) = o(f(1))$'+'  ?');
plt.legend();plt.show()
```

#### <u>Functional Iteration</u>

$lg^*\;n = min\{i \geq o: lg^{(i)}\;n \leq 1\}$

```python
def lg_star(n):
    n_p = np.log(n)/np.log(2)
    if n<=1.0: return 0.0
    else:      return 1.0+lg_star(n_p)
    
n = np.linspace(1,5,5) 
f1 = np.log(n)/np.log(2)                                 
plt.plot(n,f1,label='$f1(n)=lg\; n$')
f2 = [lg_star(n[i]) for i in range(len(n))]                          
plt.plot(n,f2,label='$f2(n)=lg^{*}\;n$')
plt.title('$f2(n) = o(f(1))$'+'  ?');
plt.legend();plt.show()
```

#### <u>Supa Asymptotic Inequality!!!</u>

$\forall n \in \mathbb{Z}^{+}, \mu,\epsilon \in \mathbb{R} \wedge \mu>0 \wedge 0 < \epsilon < 1:$

$O(1) << lg(n)^\epsilon << lg(n) << lg(n)^{1+\mu} << n^\epsilon << n << n^{1+\mu} << 2^{n^\epsilon} << 2^n << 2^{n^{1+\mu}}$
